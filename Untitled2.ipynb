```python
import zipfile
import tensorflow as tf
import os
# import  imageio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from keras.utils import to_categorical #For keras > 2.0
# import cv2
import scipy.io
from sklearn.model_selection import train_test_split
from keras import initializers
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from pyESN.pyESN import ESN
```

```python
mat = scipy.io.loadmat('kicked_data_05_no_source_decay.mat')


ns = mat['n_s']
ns_full = mat['ns_full']
Time = mat['Time']


ns = np.reshape(ns,(len(ns[0]),1))
ns_full_extended = []
output = []
for i in range(len(ns)):
    ns_full_extended.append(abs(ns_full[0][i][0]))
    output.append({abs(ns[i][0]):Time[0][i]})
```

```python
plt.plot(ns)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(ns_full_extended,output, test_size=0.3)
```

```python
Y_train_output = []
Y_train_time = []

Y_test_output = []
Y_test_time = []


for ele in Y_train:
  key = list(ele.keys())
  value = list(ele.values())
  Y_train_output.append(key)
  Y_train_time.append(value[0])
Y_train_output = np.reshape(Y_train_output,(len(Y_train_output),1))

for ele in Y_test:
  key = list(ele.keys())
  value = list(ele.values())
  Y_test_output.append(key)
  Y_test_time.append(value[0])
# Y_test_output = np.reshape(Y_test_ouput,(len(Y_test_output),1))
```

```python
# regressor = RandomForestRegressor(n_estimators=100)
regressor = ExtraTreesRegressor()
regressor.fit(X_train, Y_train_output)
Y_predict_RFR = regressor.predict(X_test)
```
